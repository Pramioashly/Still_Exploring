```python
def pd_sql(query):
    return pd.read_sql_query(query, conn)
```

```python
#LEFT JOIN
```

```python
pd_sql('''
    SELECT 
        person.id,
        person.name,
        drivers_license.id AS license_id,
        drivers_license.age
    FROM drivers_license
    LEFT JOIN person
    ON person.license_id = drivers_license.id;
''')
```

```python
#RIGHT JOIN
```

```python
pd_sql('''
    SELECT 
    person.Id,
    person.Name,
    drivers_license.id AS license_id,
    drivers_license.age
FROM drivers_license
LEFT JOIN person
ON person.license_id = drivers_license.id;

''')
```

```python
#OUTER JOIN
```

```python
pd_sql('''
    -- LEFT JOIN part (left table + matches)
SELECT 
    person.Id,
    person.Name,
    drivers_license.id AS license_id,
    drivers_license.age
FROM person
LEFT JOIN drivers_license
ON person.license_id = drivers_license.id

UNION

-- RIGHT JOIN part (right table + matches)
SELECT 
    person.Id,
    person.Name,
    drivers_license.id AS license_id,
    drivers_license.age
FROM drivers_license
LEFT JOIN person
ON person.license_id = drivers_license.id;


''')
```

```python
#INNER JOIN
```

```python
pd_sql('''
   SELECT 
    person.Id,
    person.Name,
    drivers_license.id AS license_id,
    drivers_license.age
FROM person
INNER JOIN drivers_license
ON person.license_id = drivers_license.id;

''')
```
